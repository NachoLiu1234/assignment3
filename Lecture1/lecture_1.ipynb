```python
def create_gramma(string, spliter='=', line_split='\n'):
    gramma = {}
    string = string.strip().split(line_split)
    for line in string:
        line = line.split(spliter, 1)
        gramma[line[0].strip()] = [ el.strip().split(' ') for el in line[1].split('|')]
    return gramma
        
```

```python
gramma = create_gramma(host)
gramma
```

```python
def generate(gramma, target):
    if type(target) == list:
        return ''.join([generate(gramma, el) for el in target])
    if target not in gramma:
        return target if target != 'null' else ''
    else:
        return generate(gramma, random.choice(gramma[target]))
        
```

```python
for _ in range(20):
    print(generate(gramma, 'host'))
```

```python
_1_gram_jieba_list = jieba_list
_2_gram_jieba_list = _2_gram_jieba_list
count_1_gram_jieba_list = Counter(_1_gram_jieba_list)
count_2_gram_jieba_list = Counter(_2_gram_jieba_list)
```

$$ Pr(sentence) = Pr(w_1 \cdot w_2 \cdots w_n) = \prod \frac{count(w_i, w_{i+1})}{count(w_{i+1})}$$

```python
def get_count(word, counter_dict):
    if word not in counter_dict:
        return counter_dict.most_common()[-1][-1]
    else:
        return counter_dict[word]
```

```python
get_count('今天', count_1_gram_jieba_list)
```

```python
get_count('今天晚上', count_2_gram_jieba_list)
```

```python
def get_probility(sentence):
    p = 1
    one_sentence_list = list(jieba.cut(sentence))
    two_sentence_list = [one_sentence_list[i] + one_sentence_list[i + 1] for i in range(len(one_sentence_list) - 1)]
    for i in range(len(two_sentence_list)):
        two_count = get_count(two_sentence_list[i], count_2_gram_jieba_list)
        one_count = get_count(one_sentence_list[i + 1], count_1_gram_jieba_list)
        p *= (two_count / one_count)
    return p
```

```python
two_gram_model = get_probility
```
